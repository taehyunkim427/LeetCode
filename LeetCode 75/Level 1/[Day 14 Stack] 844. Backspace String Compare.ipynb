### Backspace String Compare
---
Given two strings s and t, return true if they are equal when both are typed into empty text editors. '#' means a backspace character.

Note that after backspacing an empty text, the text will continue empty.

|Input|Output|Explanation|
|:--|:--|:--|
|s = "ab#c", t = "ad#c"|true|Both s and t become "ac".|
|s = "ab##", t = "c#d#"|true|Both s and t become "".|
|s = "a#c", t = "b"|false|s becomes "c" while t becomes "b".|
>$Constraints:$  
>- $1 <= s.length, t.length <= 200$
>- $s$ $and$ $t$ $only$ $contain$ $lowercase$ $letters$ $and$ $'#'$ $characters.$

```java
import java.time.Duration;
import java.time.Instant;
import java.util.stream.*;

class Solution {
    public boolean backspaceCompare(String s, String t) {
        StringBuffer sbs = new StringBuffer(), sbt = new StringBuffer();

        s.chars().forEach(c -> {
            if(c == '#'){
                if(sbs.length() - 1 >= 0) sbs.deleteCharAt(sbs.length()-1);
            }
            else sbs.append(String.valueOf((char)c));
        });

        t.chars().forEach(c -> {
            if(c == '#'){
                if(sbt.length() - 1 >= 0) sbt.deleteCharAt(sbt.length()-1);
            }
            else sbt.append(String.valueOf((char)c));
        });

        return sbs.toString().equals(sbt.toString()) ? true : false;
    }
}
```

```java
/* Measure execution time */
Runtime.getRuntime().gc();
Instant startTime = Instant.now();

Solution solution = new Solution();
String s = "ab#c", t = "ad#c";
boolean answer = solution.backspaceCompare(s, t);

Instant endTime = Instant.now();
answer;
```

```java
/* Measure execution space */
Runtime.getRuntime().gc();
long startSpace = Runtime.getRuntime().totalMemory() - Runtime.getRuntime().freeMemory();

Solution solution = new Solution();
String s = "ab#c", t = "ad#c";
boolean answer = solution.backspaceCompare(s, t);

long endSpace = Runtime.getRuntime().totalMemory() - Runtime.getRuntime().freeMemory();
answer;
```

```java
System.out.println((endSpace - startSpace) / 1024 / 1024 + " MB");
System.out.println(Duration.between(startTime, endTime).toMillis() + " ms");
```

### Simple and short Java solution 1ms
---

```java
class SimpleSolution {
    public boolean backspaceCompare(String s, String t) {
        return backspace(t).equals(backspace(s));
    }

    public String backspace(String s){
        StringBuilder sb = new StringBuilder(s);
        int index = 0;
        for (int i = 0; i < sb.length(); i++) {
            if (sb.charAt(i) == '#') {
                index = Math.max(i - 2, -1);
                if (i == 0) sb.delete(0, 1);
                    else sb.delete(i - 1, i + 1);
                i = index;
            }
        }
        return sb.toString();
    } 
}
```

```java
/* Measure execution time */
Runtime.getRuntime().gc();
Instant startTime = Instant.now();

SimpleSolution simpleSolution = new SimpleSolution();
String s = "ab#c", t = "ad#c";
boolean answer = simpleSolution.backspaceCompare(s, t);

Instant endTime = Instant.now();
answer;
```

```java
/* Measure execution space */
Runtime.getRuntime().gc();
long startSpace = Runtime.getRuntime().totalMemory() - Runtime.getRuntime().freeMemory();

SimpleSolution simpleSolution = new SimpleSolution();
String s = "ab#c", t = "ad#c";
boolean answer = simpleSolution.backspaceCompare(s, t);

long endSpace = Runtime.getRuntime().totalMemory() - Runtime.getRuntime().freeMemory();
answer;
```

```java
System.out.println((endSpace - startSpace) / 1024 / 1024 + " MB");
System.out.println(Duration.between(startTime, endTime).toMillis() + " ms");
```
